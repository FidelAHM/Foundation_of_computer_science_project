```python
import os
import pandas as pd
import zstandard as zstd
from io import BytesIO

extract_path = "/content/trendingYT/trendingYT"

dfs = []

for file in os.listdir(extract_path):
    if file.lower().endswith(".zst"):
        file_path = os.path.join(extract_path, file)
        print("Reading:", file)

        # Decompress .zst file
        with open(file_path, "rb") as f:
            dctx = zstd.ZstdDecompressor()
            decompressed = dctx.decompress(f.read())

        # Read CSV
        df = pd.read_csv(
            BytesIO(decompressed),
            encoding="latin1",
            on_bad_lines="skip"
        )

        # Extract country from filename (e.g. USvideos.csv.zst → US)
        country = file.split("videos")[0]
        df["country"] = country

        dfs.append(df)

# Concatenate all dataframes into one
final_df = pd.concat(dfs, ignore_index=True)

print("Final dataframe shape:", final_df.shape)
final_df.head()
```

```python
import os
import pandas as pd

categories_path = "/content/trendingYT/trendingYT"
```

```python
results = []

for file in os.listdir(categories_path):
    if file.endswith("_category_id.json"):
        file_path = os.path.join(categories_path, file)
        print("Reading:", file)

        # Extract country code (US, CA, etc.)
        country = file.split("_")[0]

        # Read JSON
        data = pd.read_json(file_path)

        # Normalize items
        categories_df = pd.json_normalize(data["items"])

        # Count non-assignable categories
        not_assignable_count = (~categories_df["snippet.assignable"]).sum()

        results.append({
            "country": country,
            "not_assignable_categories": not_assignable_count
        })
```

```python
not_assignable_by_country = pd.DataFrame(results)
not_assignable_by_country
```
